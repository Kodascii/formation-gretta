## 01. Connect to the MySQL and retrieve data

```python
import mysql.connector as mysql
from mysql.connector import Error

db_config = {
    'host': 'localhost',
    'user': 'kodascii',
    'password': 'pass123',
    'database': 'workshop',
}

iris_db = ()   # tupple of records


try:
    conn = mysql.connect(**db_config)
    if conn.is_connected():
        # Create a cursor
        cursor = conn.cursor()

        # Execute a SELECT query
        cursor.execute("SELECT * FROM iris")

        # Fetch results if the query is a SELECT
        iris_db = cursor.fetchall()

        if iris_db:
            for row in iris_db:
                print(row)
        else:
            print("No results found.")
        
except Error as e:
    print("Error while connecting to MySQL", e)
```

## 02. Create iris table (*using pandas and python*)

```python
import pandas as pd
irisData = pd.read_csv('iris.csv',index_col=False)

if conn.is_connected():
    cursor.execute('DROP TABLE IF EXISTS iris;')
    print('Creating table....')
    cursor.execute("CREATE TABLE iris (sepal_length FLOAT(2,1) NOT NULL, sepal_width FLOAT(2,1) NOT NULL, petal_length FLOAT(2,1) NOT NULL, petal_width FLOAT(2,1),species CHAR(11)NOT NULL)")
    print("iris table is created....")

    for i,row in irisData.iterrows():
            sql = "INSERT INTO workshop.iris VALUES (%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            conn.commit()
```

## 03. Query database

```python
# Execute query
sql = "SELECT * FROM iris"
cursor.execute(sql)

# Fetch all the records
result = cursor.fetchall()
for i in result:
    print(i)
```

## 04. Inserting Pandas DataFrames into a Database Using the to_sql() Function

```python
from sqlalchemy import create_engine

# create sqlalchemy engine
db_uri = f"mysql://{db_config['user']}:{db_config['password']}@localhost:3306/{db_config['database']}"
engine = create_engine(db_uri)

# Insert whole DataFrame into MySQL
irisData.to_sql('iris_copy', con=engine, if_exists='append', chunksize=1000, index=False)
```
